```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
x = np.arange(51)
y = np.arange(0,101,2)
z = x ** 3
```

```python
fig=plt.figure()
axes=fig.add_axes([0,0,1,1])
axes.plot(x,y,"Red")
axes.set_title("X vs Y Graph")
axes.set_xlabel("X values")
axes.set_ylabel("Y values")
```

```python
fig=plt.figure(figsize=(8,6))
axes1=fig.add_axes([0,0.1,0.8,0.8])
axes2=fig.add_axes([0.1,0.5,0.4,0.3])
axes1.plot(x,z,"Purple")
axes1.set_title("Outher Graph-X Vs Z Graph")
axes2.plot(x,y,"Red")
axes2.set_title("İnner Graph-X Vs Y Graph")
```

```python
fig,axes=plt.subplots(nrows=1,ncols=2,figsize=(10,6))
axes[0].plot(x,y,"Red",marker="o",markerfacecolor="Black")
axes[0].set_title("X Vs Y Graph")
axes[0].set_xlabel("X Values")
axes[0].set_ylabel("Y Values")
axes[1].plot(x,z,"purple",ls="-.",lw=3)
axes[1].set_title("X Vs Z Graph")
axes[1].set_xlabel("X Values")
axes[1].set_ylabel("Z Values")
plt.tight_layout()
```

```python
N=1000
x=np.linspace(-2*np.pi*5,2*np.pi*5,N)
y=np.linspace(-2*np.pi*5,2*np.pi*5,N)
X,Y=np.meshgrid(x,y)
R=np.sqrt(X**2+Y**2)
Z=np.cos(R)*np.exp(-R/5)
plt.figure(figsize=(8,8))
ax=plt.axes(projection="3d")
ax.plot_surface(X,Y,Z,cmap="jet")
```

```python
N=1000
x=np.linspace(-2*np.pi,2*np.pi,N)
y=np.linspace(-2*np.pi,2*np.pi,N)
X,Y=np.meshgrid(x,y)
Z=np.sin(X)*np.cos(Y)
plt.figure(figsize=(8,8))
ax=plt.axes(projection="3d")
ax.plot_surface(X,Y,Z,cmap="jet")
```
