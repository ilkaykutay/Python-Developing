```python
from statistics import mean
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style
# Grafigin stilini değiştirmek için olmasa da olur
style.use('fivethirtyeight')
```

```python
xs = np.array([5.19,5.71,6.88,7.41,8.20], dtype=np.float64)
ys = np.array([0.669,0.731,1.110,1.1,1.4], dtype=np.float64)
```

```python
m = ((mean(xs) * mean(ys)) - mean((xs * ys))) / (mean(xs)**2 - mean(xs**2))
```

```python
b = mean(ys) - (m * mean(xs))
```

```python
def regresyon(xs, ys):
    m = ((mean(xs) * mean(ys)) - mean((xs * ys))) / (mean(xs)**2 - mean(xs**2))
    b = mean(ys) - (m * mean(xs))
    return m, b
```

```python
m, b = regresyon(xs, ys)
b
```

```python
line = [(m * x) + b for x in range(0, 10)]
```

```python
plt.scatter(xs, ys,linewidths=5) # Gerçek veriyi çizmek
plt.plot(line, color='r') # Regresyon doğrusunu çizmek
plt.xlabel('Frekans(Hz)')
plt.ylabel('Volt(V)')
plt.plot()
```

```python
import numpy as np

#polynomial fit with degree = 2
model = np.poly1d(np.polyfit(xs,ys,2))

#add fitted polynomial line to scatterplot
polyline = np.linspace(1, 60, 100)
plt.scatter(xs,ys,linewidths=5)
plt.plot(polyline, model(polyline),"Red")
plt.xlabel("d(cm)")
plt.ylabel("Volt(V)")
plt.show()
```
